```python
#2024/11/30
import socket
import cv2
import pickle
import struct
import time
import threading
import serial
from queue import Queue


#    
command_queue= Queue()

#    

def send_to_arduino_thread():
    while True:
        message = command_queue.get()
        if message is None:
            break
        try:
            ser1.write((message + '\n').encode())
            ser2.write((message + '\n').encode())
            time.sleep(0.05)
        except Exception as e:
            print(f"Failed to send to Arduino: {e}")

#    
arduino_thread = threading.Thread(target=send_to_arduino_thread)
arduino_thread.start()


#   
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
host_ip = '192.168.245.209'  # server_ip
port = 9997 # server_port
socket_address = (host_ip, port)

#     
server_socket.bind(socket_address)
server_socket.listen(5)
print("Listening at:", socket_address)

#   
client_socket, addr = server_socket.accept()
print('Connection from:', addr)

#   
cap = cv2.VideoCapture(0)  # 0    .
data_num0=0
data_num1=0
data_num2=0
data_num3=0
data_num4=0
data_num5=0
data_num6=0
data_num7=0
data_num8=0
data_num9=0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    alpha = 0.9
    beta = 0
    frame = cv2.convertScaleAbs(frame, alpha=alpha, beta=beta)
    #   
    a = pickle.dumps(frame)
    message = struct.pack("Q", len(a)) + a
    #   
    client_socket.sendall(message)

    #    
    data = client_socket.recv(1024).decode()
    print("Received from client:", data)
    if '0' in data:
        print("cart start")
        send_to_arduino1('o') #command_queue1.put('o')
        time.sleep(0.01)

    if '1' in data:
        print("Command 'f' put in queue before")
        send_to_arduino1('f')
        time.sleep(0.01)
        control_Vertical_Horizontal(1)

    if '2' in data:
        data_num2 += 1
        if(data_num2 == 3):
            control_Vertical_Horizontal(4)  #LEFT
            data_num2 = 0

    if '3' in data:
        data_num3 += 1
        if(data_num3 == 3):
            control_Vertical_Horizontal(3)  # RIGHT
            data_num3 = 0

    if '4' in data:
        data_num4 += 1
        if(data_num4 == 3):
            control_Vertical_Horizontal(5)  # UP
            data_num4 = 0

    if '5' in data:
        data_num5 += 1
        if(data_num5 == 3):
            control_Vertical_Horizontal(6)  #DOWN
            data_num5 = 0

    if '6' in data:               # wait
        print("loading")

    if '7' in data:              # servo
        send_to_arduino2('s')
        time.sleep(0.01)

    if '8' in data:            # standard
        control_Vertical_Horizontal(0)
        time.sleep(0.01)

    if '9' in data:
        data_num9 += 1
        # detect_UP

        if(data_num9 == 1):
            control_Vertical_Horizontal(7)
            time.sleep(0.01)

        if(data_num9 == 2):
            control_Vertical_Horizontal(8)
            time.sleep(0.01)

        if(data_num9 == 3):
            control_Vertical_Horizontal(9)
            time.sleep(0.01)
            data_num = 0


    # 'q'  
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        break

    # 1 
    time.sleep(0.01)


#    None   
command_queue.put(None)
arduino_thread.join()

client_socket.close()
cap.release()
cv2.destroyAllWindows()
```
